```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read raw dataset
raw_days = pd.read_csv("../data/day.csv")
raw_days.head()
```

```python
# change the name of the dataframe to days
days = raw_days
```

```python
# Check the types of the columns
days.dtypes
```

```python
# Change dteday to datetime as it is an object at the moment
days = days.astype({"dteday": "datetime64"})
```

```python
# Check the type has been changed
days.dtypes
```

```python
# Check for NaN values
days.isna().sum()
```

```python
# Print the head of the days dataframe
days.head()
```

```python
# Save the cleaned file in a csv
days.to_csv("../data/days_cleaned.csv")
```

```python
# sns.set(rc={'figure.figsize':(15,15)})
plt.figure(figsize=(20,20))
sns.relplot(x="instant", y="cnt", kind="line", data=days)
```

```python
# Zoom in on the first 100 days of the curve
days_100 = days.iloc[:100]
```

```python
sns.relplot(x="instant", y="cnt", kind="line", data=days_100)
```

```python
# Zoom in on the first 200 days of the curve
days_200 = days.iloc[:200]
```

```python
sns.relplot(x="instant", y="cnt", kind="line", data=days_200)
```
